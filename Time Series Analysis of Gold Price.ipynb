## Linear Regression

```python
train = dt[dt.index.year <= 2017]
test = dt[dt.index.year > 2017]
print(train.shape)
print(test.shape)
```

```python
train["Price"].plot(figsize=(13,5), fontsize = 15)
test["Price"].plot(figsize=(13,5), fontsize = 15)
plt.grid()
plt.legend(['Training Data','Test Data'])
plt.show
```

```python
train_time = [i+1 for i in range(len (train))]
test_time = [i+len(train)+1 for i in range(len (test))]
len(train_time), len(test_time)
```

```python
LR_train = train.copy()
LR_test = test.copy()
```

```python
LR_train['time']= train_time
LR_test['time']= test_time
```

```python
lr = LinearRegression()
lr.fit(LR_train[['time']],LR_train["Price"].values)
```

```python
test_predictions_model1 = lr.predict(LR_test[["time"]])
LR_test['forecast'] = test_predictions_model1

plt.figure(figsize = (14,6))
plt.plot(train['Price'], label = "train")
plt.plot(test["Price"],label = 'test')
plt.plot(LR_test['forecast'], label = 'reg on time_test data')
plt. legend(loc = 'best')
plt.grid()
```

```python
def mape(actual, pred):
    return round((np. mean(abs(actual - pred)/ actual))*100, 2)
```

```python
mape_model1_test = mape(test['Price'].values, test_predictions_model1)
print("MAPE is %3.3f" % mape_model1_test, "%")
```

```python
results = pd.DataFrame({'Test Mape (%)': [mape_model1_test]}, index= ["RegressionOnTime"])
results
```

```python
plt.figure(figsize=(15, 9))
dt['Price'].plot()
plt.ylabel(None)
plt.xlabel(None)
plt.title("Gold Price")
plt.tight_layout()
plt.show()
```

```python
n_cols = 1
dataset = dt["Price"]
dataset = pd.DataFrame(dataset)
data = dataset.values

data.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range= (0, 1))
scaled_data = scaler.fit_transform(np.array(data))
```

```python
# 75% to Train , 25% to Test
train_size = int(len(data) * 0.75)
test_size = len(data) - train_size
print("Train Size :",train_size,"Test Size :",test_size)
```

```python
train_data = scaled_data[0:train_size, :]
train_data.shape
```

```python
# Creating a Training set with 60 time-steps
x_train = []
y_train = []
time_steps = 100
n_cols = 1

for i in range(time_steps, len(train_data)):
    x_train.append(train_data[i-time_steps:i, :n_cols])
    y_train.append(train_data[i, :n_cols])
    if i<=time_steps:
        print('X_train: ', x_train)
        print('y_train:' , y_train)
```

```python
# Convert to numpy array
x_train, y_train = np.array(x_train), np.array(y_train)
```

```python
# Reshaping the input to (n_samples, time_steps, n_feature)
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], n_cols))
```

```python
x_train.shape , y_train.shape
```

```python
#Model
```

```python
sns.set_style('whitegrid')
import matplotlib.pyplot as plt
plt.style.use("fivethirtyeight")

from keras.models import Sequential
from keras.callbacks import EarlyStopping
from keras.layers import Dense, LSTM, Dropout

from sklearn.preprocessing import MinMaxScaler
```

```python
model = Sequential([
    LSTM(50, return_sequences= True, input_shape= (x_train.shape[1], n_cols)),
    LSTM(64, return_sequences= False),
    Dense(32),
    Dense(16),
    Dense(n_cols)
])

model.compile(optimizer= 'adam', loss= 'mse' , metrics= "mean_absolute_error")
```

```python
model.summary()
```

```python
#model training
```

```python
# Fitting the LSTM to the Training set
history = model.fit(x_train, y_train, epochs=50,batch_size=32)
```

```python
#model evaluation
```

```python
plt.figure(figsize=(12, 8))
plt.plot(history.history["loss"])
plt.plot(history.history["mean_absolute_error"])
plt.legend(['Mean Squared Error','Mean Absolute Error'])
plt.title("Losses")
plt.xlabel("epochs")
plt.ylabel("loss")
plt.show()
```

```python
#prediction
```

```python
# Creating a testing set with 60 time-steps and 1 output
time_steps = 100
test_data = scaled_data[train_size - time_steps:, :]

x_test = []
y_test = []
n_cols = 1

for i in range(time_steps, len(test_data)):
    x_test.append(test_data[i-time_steps:i, 0:n_cols])
    y_test.append(test_data[i, 0:n_cols])
x_test, y_test = np.array(x_test), np.array(y_test)
x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], n_cols))
```

```python
x_test.shape , y_test.shape
```

```python
# Get Prediction
predictions = model.predict(x_test)
```

```python
#inverse predictions scaling
predictions = scaler.inverse_transform(predictions)
predictions.shape
```

```python
#root mean square error
```

```python
#inverse y_test scaling
y_test = scaler.inverse_transform(y_test)

RMSE = np.sqrt(np.mean( y_test - predictions )**2).round(2)
RMSE
```

# ARIMA

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
# Perform seasonal decomposition
result = seasonal_decompose(dt['Price'], model='additive')

# Extract components
trend = result.trend
seasonal = result.seasonal
residual = result.resid

# Plot the components
plt.figure(figsize=(10, 8))

plt.subplot(4, 1, 1)
plt.plot(dt['Price'])
plt.title('Original Data')

plt.subplot(4, 1, 2)
plt.plot(trend)
plt.title('Trend')

plt.subplot(4, 1, 3)
plt.plot(seasonal)
plt.title('Seasonal')

plt.subplot(4, 1, 4)
plt.plot(residual)
plt.title('Residual')

plt.tight_layout()
plt.show()



```

```python
# Function to test stationarity
def test_stationarity(timeseries):
    from statsmodels.tsa.stattools import adfuller
    print("Dickey-Fuller Test Results:")
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic', 'p-value', '#Lags Used', 'Number of Observations Used'])
    for key, value in dftest[4].items():
        dfoutput[f'Critical Value ({key})'] = value
    print(dfoutput)
```

```python
# Test for stationarity
test_stationarity(dt['Price'])
```

```python
# the data is not stationary, apply differencing to make it stationary (useful for ARIMA models)
# Uncomment the following line if needed:
dt['Price_diff'] = dt['Price'].diff().dropna()
```

```python
# ARIMA Model
# Define the order of the ARIMA model (p, d, q)
p, d, q = 1, 1, 1

# Create the ARIMA model
model = ARIMA(dt['Price_diff'], order=(p, d, q))

# Fit the ARIMA model to the data
model_fit = model.fit()
```

```python
# Forecast future values
forecast_steps = 36  # Change this to the number of steps you want to forecast ahead
forecast_diff = model_fit.forecast(steps=forecast_steps)
```

```python
# Create a date range for the forecasted data
forecast_dates = pd.date_range(start=dt.index[-1], periods=forecast_steps + 1, closed='right')

# Revert differencing to get the forecast on the original scale
forecast = dt['Price'].iloc[-1] + forecast_diff.cumsum()
```

```python
# Print the forecasted values
print("Forecasted Values:")
print(forecast)
```

```python
# Visualize the forecast on the original date graph
plt.figure(figsize=(10, 6))
plt.plot(dt['Price'], label='Actual Data')
plt.plot(forecast_dates, forecast, label='Forecast')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('ARIMA Forecast')
plt.legend()
plt.show()
```
